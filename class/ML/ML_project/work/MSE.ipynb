# MSE

```python
import cv2
import numpy as np

def calculate_mse(image1, image2, patch_size, num_patches):
    # Extract patches from both images
    patches_image1 = extract_patches(image1, patch_size, num_patches)
    patches_image2 = extract_patches(image2, patch_size, num_patches)

    # Calculate MSE for each pair of corresponding patches
    mse_values = [np.mean((patch1 - patch2) ** 2) for patch1, patch2 in zip(patches_image1, patches_image2)]

    # Multiply each MSE by the number of pixels in the patches
    mse_values_weighted = [mse * np.prod(patch_size) for mse in mse_values]

    # Sum all the products
    total_weighted_mse = np.sum(mse_values_weighted)

    # Divide the sum by the total number of pixels in the original images
    total_pixels_image1 = np.prod(image1.shape)
    total_pixels_image2 = np.prod(image2.shape)
    total_pixels = (total_pixels_image1 + total_pixels_image2) // 2

    # Calculate the final MSE
    final_mse = total_weighted_mse / total_pixels

    return final_mse

def extract_patches(image, patch_size, num_patches):
    patches = []
    height, width = image.shape[:2]

    for _ in range(num_patches):
        # Randomly choose patch coordinates
        y = np.random.randint(0, height - patch_size[0] + 1)
        x = np.random.randint(0, width - patch_size[1] + 1)

        # Extract patch
        patch = image[y:y+patch_size[0], x:x+patch_size[1]]

        patches.append(patch)

    return patches

# Load the images
image1 = cv2.imread('/home/biswajitrana/Downloads/Data/Noisy_Data/0003_NOISY_SRGB_010.PNG', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('/home/biswajitrana/Downloads/Data/Output_Data/KPCA/0003_NOISY_SRGB_010_KPCA_colored.jpg', cv2.IMREAD_GRAYSCALE)
# image2 = cv2.imread('/home/biswajitrana/Downloads/Data/Output_Data/wavelet/0003_NOISY_SRGB_010_wavelet_colored.jpg', cv2.IMREAD_GRAYSCALE)

# Define patch size and number of patches
patch_size = (37,25)
num_patches = 10

# Calculate MSE
mse = calculate_mse(image1, image2, patch_size, num_patches)
print("MSE:", mse)
```
