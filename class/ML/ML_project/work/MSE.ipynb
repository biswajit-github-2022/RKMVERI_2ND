```python
import numpy as np
import cv2

def calculate_ssim(image1, image2, c1=10000000, c2=30000000):
    mu_x = np.mean(image1)
    mu_y = np.mean(image2)
    sigma_x = np.var(image1)
    sigma_y = np.var(image2)
    sigma_xy = np.cov(image1.flatten(), image2.flatten())[0, 1]

    numerator = (2 * mu_x * mu_y + c1) * (2 * sigma_xy + c2)
    denominator = (mu_x ** 2 + mu_y ** 2 + c1) * (sigma_x**2 + sigma_y**2 + c2)

    return numerator / denominator

# Example usage:
image1 = image_mean
image2 = image_True

ssim = calculate_ssim(image1, image2)
print("SSIM between the images:", ssim)
```

# MSE

```python
import cv2
import numpy as np

def calculate_mse(image1, image2, patch_size, num_patches):
    # Extract patches from both images
    patches_image1 = extract_patches(image1, patch_size, num_patches)
    patches_image2 = extract_patches(image2, patch_size, num_patches)

    # Calculate MSE for each pair of corresponding patches
    mse_values = [np.mean((patch1 - patch2) ** 2) for patch1, patch2 in zip(patches_image1, patches_image2)]

    # Multiply each MSE by the number of pixels in the patches
    mse_values_weighted = [mse * np.prod(patch_size) for mse in mse_values]

    # Sum all the products
    total_weighted_mse = np.sum(mse_values_weighted)

    # Divide the sum by the total number of pixels in the original images
    total_pixels_image1 = np.prod(image1.shape)
    total_pixels_image2 = np.prod(image2.shape)
    total_pixels = (total_pixels_image1 + total_pixels_image2) // 2

    # Calculate the final MSE
    final_mse = total_weighted_mse / total_pixels

    return final_mse

def extract_patches(image, patch_size, num_patches):
    patches = []
    height, width = image.shape[:2]

    for _ in range(num_patches):
        # Randomly choose patch coordinates
        y = np.random.randint(0, height - patch_size[0] + 1)
        x = np.random.randint(0, width - patch_size[1] + 1)

        # Extract patch
        patch = image[y:y+patch_size[0], x:x+patch_size[1]]

        patches.append(patch)

    return patches

# Load the images
image1 = cv2.imread('/home/biswajitrana/Downloads/Data/Noisy_Data/0003_NOISY_SRGB_010.PNG', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('/home/biswajitrana/Downloads/Data/Output_Data/KPCA/0003_NOISY_SRGB_010_KPCA_colored.jpg', cv2.IMREAD_GRAYSCALE)
# image2 = cv2.imread('/home/biswajitrana/Downloads/Data/Output_Data/wavelet/0003_NOISY_SRGB_010_wavelet_colored.jpg', cv2.IMREAD_GRAYSCALE)

# Define patch size and number of patches
patch_size = (37,25)
num_patches = 10

# Calculate MSE
mse = calculate_mse(image1, image2, patch_size, num_patches)
print("MSE:", mse)
```

```python
import cv2
import numpy as np

def apply_mean_shift(image):
    # Convert the image to the Lab color space
    lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)

    # Perform mean shift filtering
    shifted_image = cv2.pyrMeanShiftFiltering(lab_image, 20, 40)

    # Convert the result back to BGR color space
    result_image = cv2.cvtColor(shifted_image, cv2.COLOR_Lab2BGR)

    return result_image

def calculate_mse(image1, image2):
    # Ensure both images have the same dimensions
    assert image1.shape == image2.shape, "Images must have the same dimensions for MSE calculation"

    # Calculate the Mean Squared Error (MSE)
    mse = np.mean((image1 - image2) ** 2)

    return mse

# Load the two images
image1 = cv2.imread("/home/biswajitrana/Downloads/Data/Noisy_Data/0134_NOISY_SRGB_010.PNG")
image2 = cv2.imread('/home/biswajitrana/Downloads/Data/OUTPUT/WAVELET/0134_NOISY_SRGB_010_coloured_corrected.jpg')

# Apply mean shift filtering to both images
mean_shifted_image1 = apply_mean_shift(image1)
mean_shifted_image2 = apply_mean_shift(image2)

# Calculate the MSE between the mean-shifted images
mse_value = calculate_mse(mean_shifted_image1, mean_shifted_image2)

print("Mean Squared Error (MSE) between the two mean-shifted images:", mse_value)
```
