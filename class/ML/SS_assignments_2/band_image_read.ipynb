# Finding mean and variancve of River and Non river Data

```python
river_bands_mean=np.mean(river_bands, axis=0)
river_bands_mean
```

```python
non_river_bands_mean=np.mean(non_river_bands, axis=0)
non_river_bands_mean
```

#### Non River Variance Covariance Matrix


```python
non_river_bands_variance=np.cov(non_river_bands, rowvar=False)
non_river_bands_variance
```

# MultiVariate Normal DIstribution

```python
def multivariate_normal(x,mean,variance):
    diff=x-mean
    var_inv= np.linalg.inv(variance)
    z=np.dot(np.dot((diff.T) , var_inv) , diff)
    deter=np.linalg.det(var_inv)
    const= np.sqrt((2*np.pi))**len(x)
    p_x= (1/(const*deter**(1/2)))*np.exp(-0.5*z)
    return p_x
```

#### defining function for finding distribution value for river and non_river 

```python
def p_x(x):
    p1=multivariate_normal(x,river_bands_mean,river_bands_variance)
    p2=multivariate_normal(x,non_river_bands_mean,non_river_bands_variance)
    return p1,p2
```

# Priors

### finding priors

```python
prior_non_river=non_river_bands.shape[0]/(non_river_bands.shape[0]+river_bands.shape[0])
print(prior_non_river)
prior_river=river_bands.shape[0]/(non_river_bands.shape[0]+river_bands.shape[0])
print(prior_river)
```

### Classifying

```python
classification_img=np.zeros((img.shape[0],img.shape[1]))


for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        x=image_arr1[i,j,:]
        p1,p2=p_x(x)
        if prior_river*p1>prior_non_river*p2:
            classification_img[i,j]=0
        else:
            classification_img[i,j]=255
        # print(i,j)
```
