```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=pd.read_csv('points.csv')
plt.scatter(X['x'],X['y'])
```

```python
#step function
def step(z):
    return 2 if z>0 else 1
```

```python
#with zeros
import random
random.seed(42)
w = np.array([0, 0, 0])
for i in range(10000):
    a = random.randint(0, len(X) - 1)  
    x_sample = np.array([1,X['x'][a], X['y'][a]])
    y_hat = step(np.dot(w, x_sample))
    w =w+(X['labels'][a] - y_hat)* x_sample
print(w[0], w[1:])
m=-w[1]/w[2]
b=-w[0]/w[2]
```

```python
#ploting
x_values = np.linspace(4, 6, 100)
y_values = m * x_values + b

# Plot the line
plt.scatter(X['x'],X['y'])
plt.plot(x_values, y_values, label=f'y = {m}x + {b}')
```

```python
#Randomly
A = np.random.uniform(-10, 10)
B=np.random.uniform(-5, 5)
C=np.random.uniform(5, 7)
w_random=np.array([C,A,B])
for i in range(10000):
    a = random.randint(0, len(X) - 1)  
    x_sample = np.array([1,X['x'][a], X['y'][a]])
    y_hat = step(np.dot(w_random, x_sample))
    w_random =w_random+(X['labels'][a] - y_hat)* x_sample
print(w_random[0], w_random[1:])
m=-w_random[1]/w_random[2]
b=-w_random[0]/w_random[2]
```

```python
x_values = np.linspace(4, 6, 100)
y_values = m * x_values + b

# Plot the line
plt.scatter(X['x'],X['y'])
plt.plot(x_values, y_values, label=f'y = {m}x + {b}')
```

```python
#with the help of the data set,
#using linear regression
x=np.array(X['x'])
y=np.array(X['y'])
x_bar=np.mean(x)
y_bar=np.mean(y)
m_1=np.sum((y-y_bar)*(x-x_bar))/np.sum((x-x_bar)**2)
b_1=y_bar-m_1*x_bar
w_best_fit=np.array([-b_1,-m_1,1])
for i in range(10000):
    a = random.randint(0, len(X) - 1)  
    x_sample = np.array([1,X['x'][a], X['y'][a]])
    y_hat = step(np.dot(w_best_fit, x_sample))
    w_best_fit =w_best_fit+(X['labels'][a] - y_hat)* x_sample
print(w_best_fit[0], w_best_fit[1:])
m=-w_best_fit[1]/w_best_fit[2]
b=-w_best_fit[0]/w_best_fit[2]
```

```python
x_values = np.linspace(4, 6, 100)
y_values = m* x_values + b

# Plot the line
plt.scatter(X['x'],X['y'])
plt.plot(x_values, y_values, label=f'y = {m}x + {b}')
```
