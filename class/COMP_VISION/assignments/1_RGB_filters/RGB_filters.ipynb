```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
#read image
img = cv2.imread('intersection-154782_1280.webp')
#print its shape
print('Image Dimensions :', img.shape)
print(img)
```

```python

# Split the image into its RGB channels
blue_channel, green_channel, red_channel = cv2.split(img)

print(blue_channel)
print(red_channel)
print(green_channel)

# Display the original image
plt.figure(figsize=(10, 7))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Display the Blue channel
plt.subplot(2, 2, 2)
plt.imshow(blue_channel, cmap='Blues')
plt.title('Blue Channel')

# Display the Green channel
plt.subplot(2, 2, 3)
plt.imshow(green_channel, cmap='Greens')
plt.title('Green Channel')

# Display the Red channel
plt.subplot(2, 2, 4)
plt.imshow(red_channel, cmap='Reds')
plt.title('Red Channel')

plt.show()
```

```python

# Split the image into its RGB channels
blue_channel, green_channel, red_channel = cv2.split(img)

# Save the Blue channel
cv2.imwrite('blue_channel.jpg', cv2.merge([blue_channel, np.zeros_like(green_channel), np.zeros_like(red_channel)]))

# Save the Green channel
cv2.imwrite('green_channel.jpg', cv2.merge([np.zeros_like(blue_channel), green_channel, np.zeros_like(red_channel)]))

# Save the Red channel
cv2.imwrite('red_channel.jpg', cv2.merge([np.zeros_like(blue_channel), np.zeros_like(green_channel), red_channel]))
```
